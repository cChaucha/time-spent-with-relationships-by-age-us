## Tiempo dedicado por rango de edad.

```python
plt.figure(figsize=(10, 6))

# Actividades a graficar
actividades = df.columns[3:].tolist()
colores = ['blue', 'green', 'red', 'purple', 'orange', 'cyan']

# Iterar sobre cada actividad y trazar la línea correspondiente
for actividad, color in zip(actividades, colores):
    plt.plot(df['Year'], df[actividad], label=actividad, color=color)

# Agregar etiquetas, leyendas y título
plt.xlabel('Año')
plt.ylabel('Tiempo Gastado (en horas)')
plt.title('Tiempo Gastado por Rango de Edad')
plt.legend(bbox_to_anchor=(0.5, -0.15), loc='upper center')

# Mostrar el gráfico
plt.show()
```
