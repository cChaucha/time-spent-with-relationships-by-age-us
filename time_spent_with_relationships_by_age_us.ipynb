## Tiempo dedicado por rango de edad.

```python
# Crear la figura y el eje
plt.figure(figsize=(10, 6))

# Dibujar las líneas para cada actividad
plt.plot(df['Year'], df['Time spent alone, by age of respondent (United States)'], label='Time Alone')
plt.plot(df['Year'], df['Time spent with friends, by age of respondent (United States)'], label='Time with Friends')
plt.plot(df['Year'], df['Time spent with children, by age of respondent (United States)'], label='Time with Children')
plt.plot(df['Year'], df['Time spent with with parents, siblings and other family, by age of respondent (United States)'], label='Time with Family')
plt.plot(df['Year'], df['Time spent with partner, by age of respondent (United States)'], label='Time with Partner')
plt.plot(df['Year'], df['Time spent with coworkers, by age of respondent (United States)'], label='Time with Coworkers')

# Agregar etiquetas, leyendas y título
plt.xlabel('Año')
plt.ylabel('Tiempo Gastado (en horas)')
plt.title('Tiempo Gastado por Rango de Edad')
plt.legend()

# Mostrar el gráfico
plt.show()
```
